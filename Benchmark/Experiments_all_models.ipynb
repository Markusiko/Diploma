###  I. Изначальная версия

```python
rho = np.array([[1.00,  0.10,  0.45,  0.64 * np.sqrt(32.00)],
                [0.10,  1.00, -0.35, -0.24 * np.sqrt(32.00)],
                [0.45, -0.35,  1.00,  0.14 * np.sqrt(32.00)],
                [0.64 * np.sqrt(32.00), -0.24 * np.sqrt(32.00),  0.14 * np.sqrt(32.00), 32.00]])

r = 0.9
betas = np.array([1, 1, 1])
gammas = np.array([[0.9, 0.0, 0.0], # в оригинале 0.9, но мб надо 3.3 (при 0.9 бустинг везде лучше)
                   [1.0, 1.0, 0.0],
                   [1.0, 0.0, 1.0]])
```

```python
np.random.seed(999)

run_simulations(n=100, rho=rho, r=r, betas=betas, gammas=gammas, n_simulations=100)
```

```python
np.random.seed(999)

run_simulations(n=1000, rho=rho, r=r, betas=betas, gammas=gammas, n_simulations=100)
```

```python
np.random.seed(999)

run_simulations(n=10000, rho=rho, r=r, betas=betas, gammas=gammas, n_simulations=100)
```

```python
%%time
np.random.seed(999)

res_ols, res_dmf, res_boost, res_forest = [], [], [], []

for i in range(100, 20000, 50):
    clear_output(wait=True)
    print('Текущее число наблюдений:', i)
    res = run_simulations(n=i, rho=rho, r=r, betas=betas, gammas=gammas, n_simulations=100)\
        [['beta0_MAPE', 'beta1_MAPE', 'beta2_MAPE']].values
    res_ols.append(list(res[0]))
    res_dmf.append(list(res[1]))
    res_boost.append(list(res[2]))
    res_forest.append(list(res[3]))
```

```python
fig, axs = plt.subplots(3, 1, sharex=True)
fig.suptitle("Сравнение разных подходов при разных размеров выборки", fontsize=16)

sizes = list(range(100, 20000, 50))
    
for i in range(3):
    axs[i].set_title(f"MAPE{i}, %")
    axs[i].plot(sizes, np.array(res_ols)[:, i], label='OLS')
    axs[i].plot(sizes, np.array(res_dmf)[:, i], label='DMF')
    axs[i].plot(sizes, np.array(res_boost)[:, i], label='Boosting')
    axs[i].plot(sizes, np.array(res_forest)[:, i], label='Random Forest')
    if not i:
        axs[i].legend()
    
axs[2].set_xlabel('Размер выборки')
plt.tight_layout()
plt.savefig('./pictures/boost_forest_benchmark_09.pdf', dpi=200);
```
