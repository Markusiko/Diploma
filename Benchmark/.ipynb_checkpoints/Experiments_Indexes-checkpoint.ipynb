```python
%load_ext autoreload
%autoreload 2
from utils2_Indexes import *

from IPython.display import clear_output
import matplotlib.pyplot as plt 

%matplotlib inline
%config InlineBackend.figure_format = 'retina'
plt.rcParams['figure.figsize'] = 12, 8
```

###  I. Добавляем в индекс отбора вторые степени

```python
rho = np.array([[1.00,  0.10,  0.45,  0.64 * np.sqrt(32.00)],
                [0.10,  1.00, -0.35, -0.24 * np.sqrt(32.00)],
                [0.45, -0.35,  1.00,  0.14 * np.sqrt(32.00)],
                [0.64 * np.sqrt(32.00), -0.24 * np.sqrt(32.00),  0.14 * np.sqrt(32.00), 32.00]])

r = 0.9
betas = np.array([1, 1, 1])
gammas = np.array([[1.0, 0.0, 1.0, 0.0, 1.0, 1.0],
                   [0.9, 1.0, 0.0, 1.0, 0.5, 2.0], 
                   [1.0, 1.0, 0.0, 0.8, 0.0, 0.0]])
gammas = gammas.T
```

```python
np.random.seed(999)

run_simulations(n=100, rho=rho, r=r, betas=betas, gammas=gammas, n_simulations=100)
```

```python
np.random.seed(999)

run_simulations(n=1000, rho=rho, r=r, betas=betas, gammas=gammas, n_simulations=100)
```

```python
np.random.seed(999)

run_simulations(n=10000, rho=rho, r=r, betas=betas, gammas=gammas, n_simulations=100)
```

```python
%%time
np.random.seed(999)

res_ols, res_dmf, res_boost, res_forest = [], [], [], []

for i in range(100, 20000, 50):
    clear_output(wait=True)
    print('Текущее число наблюдений:', i)
    res = run_simulations(n=i, rho=rho, r=r, betas=betas, gammas=gammas, n_simulations=100) \
        [['beta0_MAPE', 'beta1_MAPE', 'beta2_MAPE']].values
    res_ols.append(list(res[0]))
    res_dmf.append(list(res[1]))
    res_boost.append(list(res[2]))
    res_forest.append(list(res[3]))
```

```python
fig, axs = plt.subplots(3, 1, sharex=True)
fig.suptitle("Сравнение разных подходов при разных размерах выборки", fontsize=16)

sizes = list(range(100, 20000, 50))
    
for i in range(3):
    axs[i].set_title(f"MAPE{i}, %")
    axs[i].plot(sizes, np.array(res_ols)[:, i], label='OLS')
    axs[i].plot(sizes, np.array(res_dmf)[:, i], label='DMF')
    axs[i].plot(sizes, np.array(res_boost)[:, i], label='Boosting')
    axs[i].plot(sizes, np.array(res_forest)[:, i], label='Random Forest')
    if not i:
        axs[i].legend()
    
axs[2].set_xlabel('Размер выборки')
plt.tight_layout()
plt.savefig('./pictures/index_with_second_degrees/boost_forest_full.pdf', dpi=200);
```

```python
fig, axs = plt.subplots(3, 1, sharex=True)
fig.suptitle("Сравнение разных подходов при разных размерах выборки (без МНК)", fontsize=16)

sizes = list(range(100, 20000, 50))[20:]
    
for i in range(3):
    axs[i].set_title(f"MAPE{i}, %")
#     axs[i].plot(sizes, np.array(res_ols)[:, i], label='OLS')
    axs[i].plot(sizes, np.array(res_dmf)[20:, i], label='DMF')
    axs[i].plot(sizes, np.array(res_boost)[20:, i], label='Boosting')
    axs[i].plot(sizes, np.array(res_forest)[20:, i], label='Random Forest')
    if not i:
        axs[i].legend()
    
axs[2].set_xlabel('Размер выборки')
plt.tight_layout()
plt.savefig('./pictures/index_with_second_degrees/boost_forest_sub.pdf', dpi=200);
```

Кажется, получилось не очень удачно: да, наши методы выиграли DMF, но проиграли обычному МНК. Попробуем исправить это, уменьшив значимость вклада вторых степеней при отборе (сделаем соответствующие коэффициенты менее значительными). Это будет проделано в разделе 3.

Также попробуем изменить распредления регрессоров и исходя из логики, что деревьям проще работать с дискретными признаками, будем использовать факторы из Пуассоновского распределения. Это проделывается в пункте 2.

В пункте 4 все скомбинируется.
