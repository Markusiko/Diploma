```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')


pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', 20)


%matplotlib inline
%config InlineBackend.figure_format = 'retina'
plt.rcParams['figure.figsize'] = 12, 8


import statsmodels.api as sm
from statsmodels.discrete.discrete_model import MNLogit
```

```python
df_leave_england = pd.read_csv('./data/transfers_21.csv', index_col=0).rename(columns={'Age': 'age'})
df_remain_england = pd.read_csv('./data/remain_21.csv', index_col=0).rename(columns={'Age': 'age'})

df_leave_france = pd.read_csv('./data/transfers_21_france.csv', index_col=0).rename(columns={'Age': 'age'})
df_remain_france = pd.read_csv('./data/remain_21_france.csv', index_col=0).rename(columns={'Age': 'age'})

df_leave_germany = pd.read_csv('./data/transfers_21_germany.csv', index_col=0).rename(columns={'Age': 'age'})
df_remain_germany = pd.read_csv('./data/remain_21_germany.csv', index_col=0).rename(columns={'Age': 'age'})

df_leave_italy = pd.read_csv('./data/transfers_21_italy.csv', index_col=0).rename(columns={'Age': 'age'})
df_remain_italy = pd.read_csv('./data/remain_21_italy.csv', index_col=0).rename(columns={'Age': 'age'})

df_leave_spain = pd.read_csv('./data/transfers_21_spain.csv', index_col=0).rename(columns={'Age': 'age'})
df_remain_spain = pd.read_csv('./data/remain_21_spain.csv', index_col=0).rename(columns={'Age': 'age'})

df_leave_england.sample(3, random_state=999)
```

```python
dict_classes = {np.nan: 0,
                'out': 1}

dict_positions = {'Goalkeeper': 'GK', 
                  'Left-Back': 'DF', 'Centre-Back': 'DF', 'Right-Back': 'DF',
                  'Defensive Midfield': 'MF', 'Left Midfield': 'MF', 'Right Midfield': 'MF',
                  'Attacking Midfield': 'MF', 'Central Midfield': 'MF',
                  'Centre-Forward': 'FW', 'Right Winger': 'FW', 
                  'Left Winger': 'FW', 'Second Striker': 'FW'}
```

```python
import datetime as dt

dict_months = {'Jan': '01', 'Feb': '02', 'Mar': '03', 'Apr': '04', 
               'May': '05', 'Jun': '06', 'Jul': '01', 'Aug': '08', 'Sep': '09', 
               'Oct': '10', 'Nov': '11', 'Dec': '12'}

def get_date(string_date):
    if string_date == '-' or string_date is np.nan:
        return np.nan
    
    month = dict_months[string_date.split(' ')[0]]
    day = string_date.split(' ')[1].split(',')[0]
    year = string_date[-4:]
    if len(day) == 1:
        day = '0' + day
    
    return dt.datetime.strptime(f'{day}-{month}-{year}','%d-%m-%Y')
```

```python
import pycountry_convert as pc

def country_to_continent(country_name):
    if country_name in ['England', 'Wales', 'Scotland', 'Kosovo', 'Bosnia-Herzegovina', 'Northern Ireland']:
        return 'Europe'
    
    if country_name in ["Cote d'Ivoire", 'DR Congo', 'The Gambia']:
        return 'Africa'
    
    if country_name == 'Korea, South':
        return 'Asia'
    
    if country_name == 'Neukaledonien':
        return 'Oceania'

    country_alpha2 = pc.country_name_to_country_alpha2(country_name)
    country_continent_code = pc.country_alpha2_to_continent_code(country_alpha2)
    country_continent_name = pc.convert_continent_code_to_continent_name(country_continent_code)
    return country_continent_name
```

```python
players = pd.concat([df_leave_england, df_remain_england])
players['is_bundesliga'] = 0
players['is_league1'] = 0
players['is_laliga'] = 0
players['is_seriea'] = 0

players_ger = pd.concat([df_leave_germany, df_remain_germany])
players_ger['is_bundesliga'] = 1
players_ger['is_league1'] = 0
players_ger['is_laliga'] = 0
players_ger['is_seriea'] = 0

players_fra = pd.concat([df_leave_france, df_remain_france])
players_fra['is_bundesliga'] = 0
players_fra['is_league1'] = 1
players_fra['is_laliga'] = 0
players_fra['is_seriea'] = 0

players_ita = pd.concat([df_leave_italy, df_remain_italy])
players_ita['is_bundesliga'] = 0
players_ita['is_league1'] = 0
players_ita['is_laliga'] = 0
players_ita['is_seriea'] = 1

players_esp = pd.concat([df_leave_spain, df_remain_spain])
players_esp['is_bundesliga'] = 0
players_esp['is_league1'] = 0
players_esp['is_laliga'] = 1
players_esp['is_seriea'] = 0

players = pd.concat([players, players_ger, players_fra, 
                     players_ita, players_esp])

players = players.loc[~(players.fee_y.isin(['End of loanJun 30, 2022', 'End of loanAug 29, 2022', 
                                           'End of loanAug 31, 2022']))] # баг в данных

# для удобства создадим колонку со статусом игроков (остался/ушел в аренду/продан)
# и укажем стоимость проданных игроков
players['status'] = players.fee_y.map(dict_classes).fillna(2) 
players['fee'] = np.nan
players.loc[players['status'] == 2, 'fee'] = players.loc[players.status == 2, 'fee_y']
players.drop(columns='fee_y', inplace=True)
players['fee'] = players['fee'].astype(float)

players['position'] = players['position'].map(dict_positions) # переобозначим позиции

# добавляем дополнительную статистику
players['goal_per_90min'] = players.apply(lambda x: 90 * int(x.goal) / int(x.minute) 
                                        if x.minute != 0 else 0, axis=1)
players['assist_per_90min'] = players.apply(lambda x: int(x.goal) / int(x.minute) 
                                        if x.minute != 0 else 0, axis=1)
players['age2'] = players['age'] ** 2

# # добавляем информацию о рейтинге клуба
# players = players.merge(clubs[['Rk', 'Squad']], left_on='club', how='left', right_on='Squad')
# players.Rk.fillna(19, inplace=True) ### !!! баг в парсинге, надо добыть эти клубы

# работаем со временем
players['joined_club'] = players['join'].apply(get_date)
players['contract_end'] = players['contract'].apply(get_date)
players.drop(['join', 'contract'], axis=1, inplace=True)

end_of_transfers = dt.datetime.strptime('10-09-2022','%d-%m-%Y') # ?? надо обсудить (для разных лиг разные даты???)
players['contract_left'] = players['contract_end'] - end_of_transfers
players['contract_left'] = players['contract_left'].dt.days
players.loc[players.contract_left < 0, 'contract_left'] = 0 # есть футболист с контрактом закончившимся в январе 2022
players['contract_left'].fillna(np.nanmedian(players['contract_left']), inplace=True)

players['time_in_club'] = end_of_transfers - players['joined_club']
players['time_in_club'] = players['time_in_club'].dt.days
players['time_in_club'].fillna(np.nanmedian(players['time_in_club']), inplace=True)

# континент
players['continent'] = players.nationality.map(country_to_continent)
players['from_Europe'] = (players['continent'] == 'Europe') * 1
players['from_homeland'] = ((players['nationality'].isin(['England', 'Wales', 'Scotland', 'Northern Ireland']) & 
                            players['is_bundesliga'] == 0) |
                            (players['nationality'].isin(['Germany']) & 
                            players['is_bundesliga'] == 1) | 
                            (players['nationality'].isin(['France']) & 
                            players['is_league1'] == 1) |
                            (players['nationality'].isin(['Spain']) & 
                            players['is_laliga'] == 1) |
                            (players['nationality'].isin(['Italy']) & 
                            players['is_seriea'] == 1)) * 1

# стоимость по трансфермаркету
players['market'].fillna(np.nanmedian(players['market']), inplace=True)

players.head()
```

## DMF

```python
X_class, y_class = players[['age', 'age2', 'contract_left',
                            'time_in_club', 'minute', 'from_homeland']], players.status

# если добавить маркет, начентся кошмар
# X_class, y_class = players[['age', 'age2', 'contract_left',
#                             'time_in_club', 'market', from_homeland', 'Rk']], players.status

X_class = sm.add_constant(X_class)
```

```python
# MNLogit statsmodels
lr = MNLogit(y_class, X_class).fit(method='newton', maxiter=10000)

df = pd.DataFrame()
prob = lr.predict()
df['proba0'] = prob[:, 0]
df['proba1'] = prob[:, 1]
df['proba2'] = prob[:, 2]

## Лямбды
df['lambda1'] = -np.log(prob[:, 0])
df['lambda2'] = prob[:, 1] * np.log(prob[:, 1]) / (1 - prob[:, 1])
df['lambda3'] = prob[:, 2] * np.log(prob[:, 2]) / (1 - prob[:, 2])

lr.summary()
```

### Бустинг

```python
from sklearn.model_selection import KFold
from sklearn.linear_model import LinearRegression
```

```python
from catboost import CatBoostClassifier
from sklearn.model_selection import GridSearchCV

param_grid = {'iterations': [1, 10, 100, 1000], 
              'max_depth': [1, 3, 6]}

### Бустинг
## Вероятности
X, y = players.loc[players.position != 'GK', 
                   ['age', 'age2', 'contract_left',
                    'time_in_club', 'minute', 'from_homeland']], players.loc[players.position != 'GK', 'status']
boosting = CatBoostClassifier(verbose=0)

searcher = GridSearchCV(boosting, param_grid=param_grid, cv=4, n_jobs=4, refit=True,
                       scoring='balanced_accuracy')
searcher.fit(X, y)

searcher.best_score_, searcher.best_params_
```

```python
prob_boost = searcher.best_estimator_.predict_proba(X)
```

```python
## Полиномы вероятностей
# Подбираем лучший по CV (по RMSE)

df = players.loc[players.position != 'GK'].copy()
df = pd.get_dummies(df, columns=['position'], drop_first=True)

ks = np.arange(1, 7)
all_rmses = []
for k in ks:
    # генерируем степени вероятностей
    for i in range(1, k+1):
        df[f'proba0^{i}'] = prob_boost[:, 0] ** i
        df[f'proba1^{i}'] = prob_boost[:, 1] ** i
        df[f'proba2^{i}'] = prob_boost[:, 2] ** i
    needed_columns = ['age', 'age2', 'goal_per_90min', 'assist_per_90min', 'from_homeland',
                      'position_FW', 'position_MF', 'minute', 'is_bundesliga', 'is_league1', 
                      'is_seriea', 'is_laliga'] \
                   + [f'proba{category}^{power}' for category in [0, 1, 2] 
                                                 for power in range(1, i)]
    
    X, y = df.loc[~df.fee.isna(), needed_columns], df.loc[~df.fee.isna(), 'fee']
    y = np.log(y)

    # считаем RMSE по KFold
    mse_for_k = []
    kf = KFold(n_splits=10)
    for train, test in kf.split(X, y):
        X_train, y_train = X.iloc[train], y.iloc[train]
        X_test, y_test = X.iloc[test], y.iloc[test]

        lm = LinearRegression().fit(X_train, y_train)
        mse_for_k.append((y_test - lm.predict(X_test)) ** 2)

    all_rmses.append(np.sqrt(np.mean(mse_for_k)))
best_k = ks[np.argmin(all_rmses)]

df_no_nans = df.loc[~df.fee.isna()]

## Линейная модель
needed_columns = ['age', 'age2', 'goal_per_90min', 'assist_per_90min', 'from_homeland',
                      'position_FW', 'position_MF', 'minute', 'is_bundesliga', 'is_league1', 
                      'is_seriea', 'is_laliga'] \
               + [f'proba{category}^{power}' for category in [0, 1, 2] for power in range(1, best_k+1)]

y_fee = np.log(df_no_nans.fee)
X_fee = df_no_nans[needed_columns]
X_fee = sm.add_constant(X_fee)

dmf_ml = sm.OLS(y_fee, X_fee).fit()
dmf_ml.summary()

#ml_coefs.append([dmf_ml.intercept_] + list(dmf_ml.coef_[:3]))
```
